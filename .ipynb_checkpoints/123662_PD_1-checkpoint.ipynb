 (1 pkt) i narysuj wykres oszacowania liczby ùúã

```julia
using Random, StatsBase, PyPlot
Random.seed!(42)

function pi(l)
    in_circle = 0
    in_square = 0
    for i in 1:l
        x, y = rand(), rand()
        if x^2 + y^2 <= 1
            in_circle += 1
        end
        in_square += 1
    end
    4 * in_circle / in_square
end

ks = [1, 2, 3, 4, 5, 6]
ls = 10 .^ ks

# wykres szacowania pi w zale≈ºno≈õci od liczby symulacji
pi_estimates = [pi(l) for l in ls]
scatter(ls, pi_estimates)
axhline(3.14, color="red")
xlabel("Liczba symulacji")
ylabel("Szacowanie œÄ")
ticklabel_format(axis="x", style="plain", useOffset=false);
```

```julia
using PyPlot

# szacowanie pi dla 1000 symulacji dla k = 1, 2, 3, 4, 5
ks = [1, 2, 3, 4, 5]
ls = 1000 .* 10 .^ ks 
results = [[pi(l) for i in 1:1000] for l in ls]

# wykres percentyli i ≈õredniej od liczby symulacji
medians = [median(result) for result in results]
lower_quants = [quantile(result, 0.05) for result in results]
upper_quants = [quantile(result, 0.95) for result in results]
plot(ls, upper_quants)
plot(ls, medians)
plot(ls, lower_quants)
xlabel("Liczba symulacji")
ylabel("Szacowanie pi")
ticklabel_format(axis="x", style="plain", useOffset=false);  
```
